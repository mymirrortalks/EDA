```python
dataset['Date_of_Journey'].str.split('/').str[0]
```

```python
dataset['Date'] = dataset['Date_of_Journey'].str.split('/').str[0]
dataset['Month'] = dataset['Date_of_Journey'].str.split('/').str[1]
dataset['Year'] = dataset['Date_of_Journey'].str.split('/').str[2]
```

```python
dataset.head()
```

```python
dataset['Date'] = dataset['Date'].astype(int)
dataset['Month'] = dataset['Month'].astype(int)
dataset['Year'] = dataset['Year'].astype(int)
```

```python
dataset.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
dataset['Arrival_Time'].str.split(' ').str[0]
```

```python
dataset['arrivaltime_hr'] = dataset['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0]

dataset['arrivaltime_minute'] = dataset['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1]
dataset
```

```python
dataset.drop('Arrival_Time', axis=1, inplace=True)


dataset['arrivaltime_hr']= dataset['arrivaltime_hr'].astype(int)
dataset['arrivaltime_minute']= dataset['arrivaltime_minute'].astype(int)
```

```python
dataset.head(3)
```

```python
dataset['Dept_hour']=dataset['Dep_Time'].str.split(':').str[0]
dataset['Dept_min']=dataset['Dep_Time'].str.split(':').str[1]
dataset['Dept_hour']=dataset['Dept_hour'].astype(int)
dataset['Dept_min']=dataset['Dept_min'].astype(int)
dataset.drop('Dep_Time',axis=1,inplace=True)
dataset.head()
```

```python
dataset.info()
```

```python
dataset['hrminute'] = dataset['Duration'].str.split(' ').str[0].str[0:-1]
dataset['hrminute'] = dataset['hrminute'].astype(int)
dataset['hrminute'] = dataset['hrminute']*60
dataset['min'] = dataset['Duration'].str.split(' ').str[1].str[0:-1]
dataset['min'] =dataset['min'].fillna(0)
dataset['min'] = dataset['min'].astype(int)
dataset.head()
```

```python
dataset['Duration_minute'] = dataset['hrminute']+dataset['min']
dataset.head()
```

```python
dataset.drop('Duration', axis=1,inplace=True)
dataset.drop('hrminute', axis=1,inplace=True)
dataset.drop('min', axis=1,inplace=True)
dataset.head()
```

```python
dataset['Total_Stops'].unique()
```

```python
dataset['Total_Stops'].mode()
```

```python
dataset['Total_Stops'] = dataset['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
dataset.drop('Route', axis=1, inplace=True)
```
